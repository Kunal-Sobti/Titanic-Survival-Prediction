```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/Kunal/OneDrive/Desktop/Coding Ninjas Machine Learning/Logistic Regression/0000000000002429_training_titanic_x_y_train.csv")
titanic=df.copy()
titanic.drop('Name',axis=1,inplace =True)
def f(s):
    if s=="male":
        return 0
    else:
        return 1
    
titanic["Sex"]=titanic.Sex.apply(f)
del df["Sex"]
titanic.drop('Ticket',axis=1,inplace =True)
titanic.drop('Cabin',axis=1,inplace =True)
titanic.Sex.fillna(titanic.Sex.median(),inplace=True)
titanic.Age.fillna(titanic.Age.median(),inplace=True)
titanic.Fare.fillna(titanic.Fare.median(),inplace=True)
titanic.SibSp.fillna(titanic.SibSp.median(),inplace=True)
titanic.Embarked.fillna(titanic.Embarked.mode(),inplace=True)
titanic = pd.get_dummies(titanic, columns = ['Sex']) #one_hot_encoded_data
titanic=pd.get_dummies(titanic, columns = ['Embarked']) #one_hot_encoded_data
titanic=titanic[[c for c in titanic if c not in ['Survived']] + ['Survived']]#moving embarked column to last
titanic_ds=np.array(titanic)
titanic_ds.shape
```

```python
clf=LogisticRegression()
clf.fit(titanic_ds[:,:10],titanic_ds[:,10])
```

```python
test_data=pd.read_csv("C:/Users/Kunal/OneDrive/Desktop/Coding Ninjas Machine Learning/Logistic Regression/0000000000002429_test_titanic_x_test (6).csv")
test_data.drop('Name',axis=1,inplace =True)
def f(s):
    if s=="male":
        return 0
    else:
        return 1
test_data["Gender"]=test_data.Sex.apply(f)
del test_data["Sex"]
test_data.drop('Ticket',axis=1,inplace =True)
test_data.drop('Cabin',axis=1,inplace =True)
test_data.Gender.fillna(test_data.Gender.median(),inplace=True)
test_data.Age.fillna(test_data.Age.median(),inplace=True)
test_data.Fare.fillna(test_data.Fare.median(),inplace=True)
test_data.SibSp.fillna(test_data.SibSp.median(),inplace=True)
test_data.Embarked.fillna(test_data.Embarked.mode(),inplace=True)
test_data = pd.get_dummies(test_data, columns = ['Gender']) #one_hot_encoded_data
test_data=pd.get_dummies(test_data, columns = ['Embarked']) #one_hot_encoded_data

titanic_test_ds=np.array(test_data)
titanic_test_ds.shape
predictions=clf.predict(titanic_test_ds)
df=pd.DataFrame(predictions)
df[0].to_csv("C:/Users/Kunal/OneDrive/Desktop/Coding Ninjas Machine Learning/Logistic Regression/predictions_new.csv",index=False)
```
